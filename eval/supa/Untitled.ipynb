```python
import pandas as pd
```

```python
excel_data = pd.read_excel('Pre Evaluation (Responses).xlsx')
pre_data = pd.DataFrame(excel_data)
pre_data = pre_data.iloc[:, 1:]

excel_data = pd.read_excel('Post Evaluation (Responses).xlsx')
post_data = pd.DataFrame(excel_data)
post_data = post_data.iloc[:, 1:]
```

```python

eval_new_cols = [
    "Your payment has been declined",
    "John give us another chance? We've got thousands of new TV shows and movies.",
    "Your number has been updated.",
    "Update your payment details",
    "Please update your payment details",
    "Free Netflix for 30 days",
    "We are sorry to say goodbye",
    "Your cancellation confirmation",
    "Account suspended",
    "New sign-in to Netflix",
    "Your account has been compromised",
    "Spreadsheet",
]
answer = [
    "Yes",
    "No",
    "No",
    "No",
    "Yes",
    "Yes",
    "No",
    "Yes",
    "Yes",
    "Yes",
    "Yes",
    "Yes"
]

for i in range(11):
    eval = pd.DataFrame({'subject': [], "real": [], "pre": [], "post": []})

    pre_true_pos = 0
    pre_true_neg =0
    pre_false_pos = 0
    pre_false_neg = 0
    pre_maybe = 0
    
    post_true_pos = 0
    post_true_neg =0
    post_false_pos = 0
    post_false_neg = 0
    post_maybe = 0
    
    assert len(answer) == len(eval_new_cols)
    for index,c in enumerate(eval_new_cols):
        assert pre_data.at[i, "Email Address"] == post_data.at[i, "Email Address"]
       
        ans = pre_data.at[i, f"Subject: {c}"]
        ans_post = post_data.at[i, f"Subject: {c}"]

        if ans == "Need more info":
            ans = "Maybe"

        if ans_post == "Need more info":
            ans_post = "Maybe"
        eval.loc[eval.shape[0]] = [c, answer[index],ans, ans_post]
        
        # pre_true = [x for x in answer if ans==x]
        # print(pre_data.at[i, "Email Address"])
        # print(pre_true)
        # print(eval)
        # ans = pre_data.at[i, f"Email Address"]
        # eval.to_csv(f"{ans.split('@')[0]}.csv", index=False)
        
        if answer[index] == "Yes" and ans == "Yes":
            # print(f"{answer[index]},{ans},{index}")
            pre_true_pos += 1

        if answer[index] == "Yes" and ans == "No":
            # print(f"{answer[index]},{ans},{index}")
            pre_true_neg += 1

        if answer[index] == "No" and ans == "No":
            # print(f"{answer[index]},{ans},{index}")
            pre_false_pos += 1

        if answer[index] == "No" and ans == "Yes":
            # print(f"{answer[index]},{ans},{index}")
            pre_false_neg += 1

        if ans == "Maybe":
            pre_maybe += 1
            
        if answer[index] == "Yes" and ans_post == "Yes":
                # print(f"{answer[index]},{ans},{index}")
            post_true_pos += 1

        if answer[index] == "Yes" and ans_post == "No":
            # print(f"{answer[index]},{ans},{index}")
            post_true_neg += 1

        if answer[index] == "No" and ans_post == "No":
            # print(f"{answer[index]},{ans},{index}")
            post_false_pos += 1

        if answer[index] == "No" and ans_post == "Yes":
            # print(f"{answer[index]},{ans},{index}")
            post_false_neg += 1

        if ans_post == "Maybe":
            post_maybe += 1

    
    print(pre_data.at[i,f"Email Address"])
    print(f"Pre: {pre_true_pos},{pre_true_neg},{pre_false_pos},{pre_false_neg},{pre_maybe}")
    print(f"Post: {post_true_pos},{post_true_neg},{post_false_pos},{post_false_neg},{post_maybe}")
    print("\n")
```
